```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Histogram

- HUE : It splits your data into groups and shows them with different colors.

```python
penguins_data = pd.read_csv('penguins_size.csv')
penguins_data[['species' , 'culmen_length_mm']]

plt.figure(figsize=(48,24))
ax = sns.histplot(data=penguins_data , x=penguins_data['culmen_length_mm'] )
ax.set_title('Column Length Distribution' , fontsize = 80)
ax.set_xlabel('Culmen Length' , fontsize = 50)
ax.set_ylabel('Frequency' , fontsize = 55)
plt.show()
```

```python
penguins_data = pd.read_csv('penguins_size.csv')
penguins_data[['species' , 'culmen_length_mm']]

plt.figure(figsize=(48,24))
ax = sns.histplot(data=penguins_data , x=penguins_data['culmen_length_mm'] ,hue=penguins_data['species']) #filter by species
ax.set_title('Column Length Distribution' , fontsize = 80)
ax.set_xlabel('Culmen Length' , fontsize = 50)
ax.set_ylabel('Frequency' , fontsize = 55)
plt.show()
```

### Box Plot

- Min = Q1 - 1.5 * IQR
- Max = Q3 + 1.5 * IQR
- Any Value greater or less than Min and Max are Outliers
- Box Plot Used to find Outliers

```python
housing_data = pd.read_csv('HousingPricesData.csv')


plt.figure(figsize=(48,24))
ax = sns.boxplot(data=housing_data , x = housing_data['Price'],hue=penguins_data['species'])
ax.set_xlabel('Price' , fontsize = 50)
ax.set_ylabel('Frequency' , fontsize = 50)
ax.set_title('Box Plot Of House Prices' , fontsize = 70)
plt.show()
```

### Violin Plot
- Min = Q1 - 1.5 * IQR
- Max = Q3 + 1.5 * IQR
- Qlmost same as box

```python
housing_data = pd.read_csv('HousingPricesData.csv')
housing_data = housing_data[['Zip' , 'Price' , 'Area' , 'Room']]

plt.figure(figsize=(48,24))
ax = sns.violinplot(data=housing_data , x = housing_data['Price'],hue=penguins_data['species'])
ax.set_xlabel('Price' , fontsize = 50)
ax.set_ylabel('Frequency' , fontsize = 50)
ax.set_title('Violin Plot Of House Prices' , fontsize = 70)
plt.show()
```

### Count Plot

```python
# SET HUE = SPECIES

penguins_data = pd.read_csv('penguins_size.csv')
plt.figure(figsize=(48,24))
ax = sns.countplot(data = penguins_data , x=penguins_data['species'] , hue=penguins_data['species'])
ax.set_xlabel('Species' , fontsize = 50)
ax.set_ylabel('Count' , fontsize = 50)
plt.show()
```

```python
# SET HUE = SEX

penguins_data = pd.read_csv('penguins_size.csv')
plt.figure(figsize=(48,24))
ax = sns.countplot(data = penguins_data , x=penguins_data['species'] , hue=penguins_data['sex'])
ax.set_xlabel('Species' , fontsize = 50)
ax.set_ylabel('Count' , fontsize = 50)
plt.show()
```

### Pie Chart

```python
penguins_data = pd.read_csv('penguins_size.csv')
plt.figure(figsize=(34,17))

plt.pie(penguins_data['species'].value_counts(), labels=penguins_data['species'].unique(),  colors = ['g' , 'b' , 'orange'] )

plt.title('Proportion Of Penguin Species' , fontsize = 50)
plt.show()
```
