```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing_data = pd.read_csv('HousingPricesData.csv')
housing_data.head()
```

```python
housing_data = housing_data[['Zip' , 'Price' , 'Area' , 'Room']]
housing_data.head()

# Adding Column 

housing_data['PricePerSqm'] = housing_data['Price'] / housing_data['Area']
print(housing_data.head())

```

### PLOTTING

####  Bar Chart

```python
# SORTED DATA

sorted_data = housing_data.sort_values('Price' , ascending = False)

x = sorted_data['Zip'][0:10]
y = sorted_data['Price'][0:10]


plt.figure(figsize = (48,24))
plt.bar(x,y , color='green' )
plt.title('Top 10 Most Expensive Houses', fontsize=56 , color = 'blue')
plt.xlabel('Zip Code' , fontsize=56 , color = 'blue')
plt.ylabel('Price in $' , fontsize=56 , color = 'blue')
plt.xticks(fontsize = 30 , rotation = 45)
plt.yticks(fontsize = 30)
plt.legend(['Price'] ,fontsize=50)
plt.show()
```

```python
# 2 bars in one bar

sorted_data = housing_data.sort_values('Price' , ascending = False)

x = sorted_data['Zip'][0:10]
y = sorted_data['Price'][0:10]
y1 = sorted_data['PricePerSqm'][0:10]*10  # chudataniki chinnaga undhi so Multiplied---> Amplified by 10

plt.figure(figsize = (48,24))
plt.bar(x,y , color='green' )
plt.plot(x,y1 , color = 'blue')
plt.title('Top 10 Most Expensive Houses', fontsize=56 , color = 'blue')
plt.xlabel('Zip Code' , fontsize=56 , color = 'blue')
plt.ylabel('Price in $' , fontsize=56 , color = 'blue')
plt.xticks(fontsize = 30 , rotation = 45)
plt.yticks(fontsize = 30)
plt.legend(['Price', 'PricePerSqm']  ,fontsize=50)

plt.show()
```

```python
# 2 bars in one bar

sorted_data = housing_data.sort_values('Price' , ascending = False)

x = sorted_data['Zip'][0:10]
y = sorted_data['Price'][0:10]
y1 = sorted_data['PricePerSqm'][0:10]*10  # chudataniki chinnaga undhi so Multiplied ---> Amplified by 10

plt.figure(figsize = (48,24))
plt.bar(x,y , color='green' )
plt.plot(x,y1 , color = 'blue' ,marker='o', markersize=30 , linewidth=10)
plt.title('Top 10 Most Expensive Houses', fontsize=56 , color = 'blue')
plt.xlabel('Zip Code' , fontsize=56 , color = 'blue')
plt.ylabel('Price in $' , fontsize=56 , color = 'blue')
plt.xticks(fontsize = 30 , rotation = 45)
plt.yticks(fontsize = 30)
plt.legend(['Price', 'PricePerSqm']  ,fontsize=50)

plt.show()
```

## Sub Plots

```python
x = sorted_data['Zip'][0:10]
y = sorted_data['Price'][0:10]
y1 = sorted_data['PricePerSqm'][0:10]

plt.subplots(figsize = (48,24))

plt.subplot(1,2,1) # 1 row : 2 columns : no.of charts we want to insert
plt.bar(x,y,color='green')

plt.subplot(1,2,2)
plt.bar(x,y1,color='blue' )
plt.show()
```
