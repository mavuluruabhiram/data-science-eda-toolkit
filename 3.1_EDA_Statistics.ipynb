```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
covid_data = pd.read_csv('covid-data.csv')

# selecting req columns 
covid_data = covid_data[['iso_code' , 'continent' , 'location' , 'date', 'total_cases' , 'new_cases']]

# MEAN
mean_value = np.mean(covid_data['new_cases'])
print('Mean Of New Cases : ' , mean_value)

#MEDIAN
median_value = np.median(covid_data['new_cases'])
print('Median Of New Cases : ' , median_value)

#MODE
mode_value = stats.mode(covid_data['new_cases'])  # numpy dont have Mode Function so we used STATS
print('Mode Of New Cases : ' , mode_value)
print('Mode Of New Cases : ' , mode_value[0]) # since MODE is collection of values more than 1 , so for 1st mode use INDEXING
#VARIANCE
variance = np.var(covid_data['new_cases'])  
print('Variance Of New Cases : ' , variance)

#STANDARD DEVIATION
standard_deviation = np.std(covid_data['new_cases'])  
print('standard Deviation Of New Cases : ' , standard_deviation)

#RANGE
min_value = np.min(covid_data['new_cases'])
max_value = np.max(covid_data['new_cases'])
range = max_value - min_value  
print('Range Of New Cases : ' , range)

#PERCENTILE
percentile_60 = np.percentile(covid_data['new_cases'],60)  
print('60th percentile Of New Cases : ' , percentile_60)

percentile_40 = np.percentile(covid_data['new_cases'],40)  
print('40th percentile Of New Cases : ' , percentile_40)

percentile_80 = np.percentile(covid_data['new_cases'],80)  
print('80th percentile Of New Cases : ' , percentile_80)

#QUARTILES
Q1 = np.percentile(covid_data['new_cases'],25) 
Q3 = np.percentile(covid_data['new_cases'],75) 
print('Quartiles Of New Cases: , Q1=' ,Q1 , 'Q3=',Q3)

#InterQuartile Range
iqr = stats.iqr(covid_data['new_cases'])
print('Inter Quartile Range Of New CAses :' , iqr)
```
