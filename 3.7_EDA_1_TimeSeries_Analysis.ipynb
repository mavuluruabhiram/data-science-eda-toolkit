```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
date_rng= pd.date_range( start='2018-1-1' , end='2023-12-31' , freq = 'ME') 
# freq = ME(month) , W (week) , D(day)
print(date_rng)
```

```python
sales = 200 + 20 *np.sin(2*np.pi + date_rng.month/12) + np.random.normal(0,5,len(date_rng))
print(sales)
```

### Creating A DataFrame

```python
df = pd.DataFrame({'Date' : date_rng , 'sales' : sales})
print(df)
```

```python
df = df.set_index('Date')
print(df)
```

### Plotting - Line Plot
- Used for Trend and Seasonality

```python
sns.lineplot(data = df , x=df.index ,y = 'sales' )
plt.show()
```

```python
trend = np.linspace(100,300, len(date_rng)) # 100 to 300 spliutted into 72 times
trend
```

```python
seasonality = 30 * np.sin(2 * np.pi *date_rng.month / 12)  # cyclic order
seasonality
```

```python
noise = np.random.normal(0,8,len(date_rng))
noise
```

```python
sales = trend + seasonality + noise
sales
```

```python
# changed Sales

sales = trend + seasonality + noise
df = pd.DataFrame({'Date' : date_rng , 'sales' : sales})
df=df.set_index('Date')
df
```

```python
sns.lineplot(data=df , x=df.index , y='sales')
```

#### To Understand The Trend , We Need To Identify Moving Mean

```python
df['RollingMean'] = df['sales'].rolling(window=12, center=True).mean()
df.head(30)

# window = how many periods to include in each rolling calculation.
```

```python
sns.lineplot(data=df , x= df.index , y='sales')
sns.lineplot(data=df , x= df.index , y='RollingMean')
```

### Checking Seasonality By Month - by grouping and averaging date by month

```python
df['Month'] = df.index.month
monthly_avg = df.groupby('Month')['sales'].mean()
```

```python
sns.lineplot(x=monthly_avg.index , y=monthly_avg.values , marker='o')
```

### Cyclic Period - Yearly

```python
df['Year'] = df.index.year
yearly_avg = df.groupby('Year')['sales'].mean()
sns.lineplot(x=yearly_avg.index , y=yearly_avg.values , marker='o')
```

### Spotting Random Or Irregular Variations

```python
# NULL VALUES

df['Detrended'] = df['sales'] - df['RollingMean']
df['Detrended']
```

### Decomposition Models

```python
pip install statsmodels
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition = seasonal_decompose( df['sales'] , model='additive , period = 12')  # here 12 is mnths , if 7 it is weeks
decomposition.plot()
plt.suptitle( ' Additive Decomposition Of Monthly Sales' , fontsize = 14)
```

```python
decomposition = seasonal_decompose( df['sales'] , model='multiplicative , period = 12')  # here 12 is mnths , if 7 it is weeks
decomposition.plot()
plt.suptitle( ' Multiplicative Decomposition Of Monthly Sales' , fontsize = 14)
```

```python
df['Trend'] = decomposition.trend
df['Seasonal'] = decomposition.seasonal
df['Residual'] = decomposition.resid

df[['sales','Trend','Seasonal','Residual']]
```

```python
df['Month'] = df.index.month
seasonal_pattern = df.groupby('Month')['Seasonal'].mean()
plt.plot(seasonal_pattern.index,seasonal_pattern.values , marker = 'o')
plt.show()
```

### Moving Averages

```python
df['SMA_12'] = df['sales'].rolling(window=12).mean()
df['SMA_6'] = df['sales'].rolling(window=6).mean()
df['SMA_3'] = df['sales'].rolling(window=3).mean()

sns.lineplot(data = df , x = df.index , y='sales' , color='black')
sns.lineplot(data = df , x = df.index , y='SMA_12' , color='red')
sns.lineplot(data = df , x = df.index , y='SMA_6' , color='blue')
sns.lineplot(data = df , x = df.index , y='SMA_3' , color='green')
plt.legend()
```

### Exponential Weighted Moving Averages

```python
#alpha = 0.3

df['ES_0.3'] = df['sales'].ewm(alpha = 0.3 , adjust=False).mean()
df['ES_0.5'] = df['sales'].ewm(alpha = 0.5 , adjust=False).mean()
df['ES_0.8'] = df['sales'].ewm(alpha = 0.8 , adjust=False).mean()
df['ES_1.0'] = df['sales'].ewm(alpha = 1.0 , adjust=False).mean()

sns.lineplot(data = df , x = df.index , y='sales' ,color='black') # applpha =1 ,That means the smoothed series is exactly equal to the original sales column.
sns.lineplot(data = df , x = df.index , y='ES_0.3' ,color='orange')
sns.lineplot(data = df , x = df.index , y='ES_0.5' , color='red')
sns.lineplot(data = df , x = df.index , y='ES_0.8' , color='blue')
sns.lineplot(data = df , x = df.index , y='ES_1.0' , color='green')
```

### Simple Exponential Smooting

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
model = SimpleExpSmoothing(df["sales"])
fit = model.fit(smoothing_level = 0.3, optimized = False)
df["SES_0.3"] = fit.fittedvalues

# plotting is pending
```
