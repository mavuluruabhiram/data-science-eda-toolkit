```python
import numpy as np 
import pandas as pd 
```

```python
# Selecting 

chemistry_2019 = nobel.loc[0, "prizes"]
chemistry_2019
```

```python
pd.json_normalize(data=chemistry_2019)
```

```python
pd.json_normalize(data=chemistry_2019, record_path="laureates")
```

```python
pd.json_normalize(data=chemistry_2019, 
                  record_path="laureates",
                  meta=["year", "category"])
```

```python
winners = pd.json_normalize(data=chemistry_2019, 
                  record_path="laureates",
                  meta=["year", "category"])
type(winners)
```

```python
winners
```

```python
def add_laureates_key(entry):
    entry.setdefault("laureates", [])
nobel["prizes"].apply(add_laureates_key)
```

```python
full_winners = pd.json_normalize(
    data=nobel["prizes"],
    record_path="laureates",
    meta=["year", "category"]
)
full_winners
```
