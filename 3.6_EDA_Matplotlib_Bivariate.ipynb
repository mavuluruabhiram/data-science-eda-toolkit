### CrossTab

- Numerical   to Numerical
- Numerical   to Categorical
- Categorical to Categorical 

- CrossTab dont need any plotting libraries
- We hhave to specify 2 parameters : 
    - Index - Row Labels
    - Columns - column labels
    - Aggregate func - numpy func ( mean.....et)

```python
penguins_data = pd.read_csv('penguins_size.csv')
crosstab = pd.crosstab(index=penguins_data['species'] , columns=penguins_data['sex'] )
print(crosstab)
```

```python
penguins_data = pd.read_csv('penguins_size.csv')
pivot = pd.pivot_table(data=penguins_data, index=['species' , 'sex' ], values=['culmen_length_mm' , 'body_mass_g' ] , aggfunc=np.mean )
print(pivot)
```

```python
penguins_data = pd.read_csv('penguins_size.csv')
pivot = pd.pivot_table(data=penguins_data, index='species' , columns='sex', values=['culmen_length_mm' , 'body_mass_g' ] , aggfunc=np.mean )
print(pivot)
```

# Box Plot

```python
penguins_data = pd.read_csv('penguins_size.csv')


plt.figure(figsize=(10,6))
ax = sns.boxplot(data=penguins_data , x = penguins_data['species'], y = penguins_data['culmen_length_mm'] )
ax.set_xlabel('Species' , fontsize = 20)
ax.set_ylabel('Culmen' , fontsize = 20)
ax.set_title('Box Plot ' , fontsize = 30)
plt.show()
```

### Histogram

```python
penguins_data = pd.read_csv('penguins_size.csv')

penguins_male_data = penguins_data[penguins_data['sex']=='MALE']
penguins_female_data = penguins_data[penguins_data['sex']=='FEMALE']

plt.figure(figsize=(30,15))
sns.histplot(data=penguins_male_data , x='culmen_length_mm',alpha=0.5 , color='green')
sns.histplot(data=penguins_female_data , x='culmen_length_mm' ,alpha = 0.5 , color = 'blue')

plt.show()
```

 ## CORRELATION

 - Calories --------------------------- Weight
    -    X     ---------------------------   Y
    -   f(X)   ---------------------------   Y

 - Positive Correlation ( > 0 and <=1 ) ( Temperature --------- Ice Cream Sales)
 - Negitive Correlation ( >= -1 and < 0) (  Exercise ------- Weight  )
 - No Correlation ( 0 )( Speed ---- Petrol )

```python
penguins_data = pd.read_csv('penguins_size.csv')
penguins_data = penguins_data[['culmen_length_mm' , 'body_mass_g']]
penguins_corr = penguins_data.corr()

sns.heatmap(penguins_corr , vmin=-1 , vmax=1 , annot=True)  # annot = information
plt.show()
```
