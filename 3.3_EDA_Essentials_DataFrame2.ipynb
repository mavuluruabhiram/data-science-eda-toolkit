```python
import pandas as pd
import numpy as np
```

```python
marketing_data = pd.read_csv('marketing_campaign.csv')
marketing_data[['ID' , 'Year_Birth' , 'Education' , 'Marital_Status' , 'Income' , 'Kidhome' , 'Teenhome' , 'Dt_Customer' , 'Recency' , 'NumStorePurchases' , 'NumWebVisitsMonth']]
```

### Sorting

```python
# SORTING

marketing_data = marketing_data.sort_values('NumStorePurchases' , ascending=False)
marketing_data
```

### CUTTING DATA - CUT - bucketing


```python
bucketted_data = pd.cut(marketing_data['NumStorePurchases'] , bins = [0,4,8,13] , labels = ['Low' , 'Moderate' , 'High'])
print(bucketted_data.head())
```

### Drop Duplicates

```python
marketing_data2 = marketing_data[['Education' , 'Marital_Status' , 'Kidhome' , 'Teenhome']]
marketing_data2.drop_duplicates()
```

### DROPPING

```python

marketing_data2 = marketing_data2.drop(labels=[499]) # dropping row 

marketing_data2 = marketing_data2.drop(labels=['Education'],axis=1) # dropping column
print(marketing_data2)
```
